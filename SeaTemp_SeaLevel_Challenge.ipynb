```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SeaTemp_SeaLevel_anomaly.csv")
```

```python
year = df["Year"]
temp = df["Sea temperature anomaly (C)"]
mean = df["Change in mean sea level (mm)"]
```

```python
plt.plot(year, temp, color = "blue", label = "Sea Temp Anomaly")
plt.ylabel("Sea Temp (C)")
plt.legend()
plt.twinx()
plt.plot(year, mean, color = "red", label = "Change in Mean")
plt.ylabel("Change in Mean (mm)")
plt.legend(loc="lower right")
plt.title("Change in mean of Sea Temp")
plt.xlabel("Years")
plt.show()
```
